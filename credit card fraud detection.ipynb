```python
x = df.drop(columns="Class",axis=1)
y = df["Class"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,stratify=y,test_size=0.7,random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
p1 = Pipeline([("scaler1",StandardScaler()),("lin_reg",LinearRegression())])
p2 = Pipeline([("rand",RandomForestClassifier())])
p3 = Pipeline([("grad",GradientBoostingClassifier())])
p4 = Pipeline([("decis",DecisionTreeClassifier())])
p5 = Pipeline([("scaler2",StandardScaler()),("log_reg",LogisticRegression())])
```

```python
pipelines = [p1,p2,p3,p4,p5]
pipelines
```

```python
p1.fit(x_train,y_train)
p1.score(x_test,y_test)*100
```

```python
p2.fit(x_train,y_train)
p2.score(x_test,y_test)*100
```

```python
p3.fit(x_train,y_train)
p3.score(x_test,y_test)*100
```

```python
p4.fit(x_train,y_train)
p4.score(x_test,y_test)*100
```

```python
p5.fit(x_train,y_train)
p5.score(x_test,y_test)*100
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```
